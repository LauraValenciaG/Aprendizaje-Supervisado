### 1. ¿Qué son las Redes Neuronales?

Imagina que tienes un cerebro lleno de pequeñas piezas llamadas neuronas. Cada neurona toma una pequeña decisión, como “¿es esta cosa es redonda o cuadrada?” y luego pasa la respuesta a la siguiente neurona. Si muchas neuronas trabajan juntas, pueden resolver problemas más complejos, como identificar si un animal es un gato o un perro. Así es como las redes neuronales “aprenden” de los ejemplos, viendo patrones y tomando decisiones. Es como si fueran pequeños detectives que colaboran para encontrar respuestas.

### 2. Tipos de Redes Neuronales:

Cada tipo de red neuronal es como una herramienta especial para resolver diferentes tipos de problemas. Vamos a ver cada una:

**1. Red Red Neuronal Artificial (ANN):**

Piensa en las redes ANN como las más básicas, conectadas de una manera muy simple, donde la información fluye desde un inicio hasta un final sin dar vueltas. Esta red es buena para tareas sencillas, como adivinar si una foto es de un juguete grande o pequeño, o si algo es caliente o frío. Son las primeras redes neuronales que se inventaron.

**2. Red Neuronal Convolucional (CNN):**

Las redes CNN están diseñadas especialmente para ver imágenes. Imagínate que usan “lupas” para ver detalles en diferentes partes de una foto. Primero miran pedacitos pequeños, y después juntan todo para entender la imagen completa. Son perfectas para identificar cosas en fotos, como el rostro de alguien o un animal, y son muy utilizadas en aplicaciones como filtros de fotos o reconocimiento facial

**3. Red Neuronal Recurrente (RNN):**

Las RNN funcionan bien cuando necesitamos recordar lo que pasó antes, como cuando escuchamos una canción o leemos una historia. Piensa en esta red como un grupo de amigos que pasa un mensaje por una fila, pero cada amigo se acuerda de lo que le dijeron antes. Estas redes son ideales para trabajar con palabras, música o cualquier cosa que tenga un orden en el tiempo. Se utilizan en traducción de idiomas, generación de texto o incluso para predecir qué palabra vendrá después en una frase.

**4. Autoencoders:**

Un Autoencoder es una red que toma algo grande, lo guarda en una forma más pequeña y luego trata de reconstruirlo igualito al original. Imagina que tienes una caja de juguetes y quieres llevarla contigo. En lugar de llevar la caja entera, decides meter solo los juguetes más importantes en una bolsa. Cuando llegas a otro lugar, vuelves a armar todo usando solo los juguetes de la bolsa. Esto ayuda a reducir el espacio necesario sin perder la información clave, y es muy útil para comprimir datos, como imágenes.

**5. Transformers:**

Transformers son como una versión muy avanzada de las redes. Imagina que tienes un amigo que puede leer un libro y recordar el inicio, aunque esté leyendo el final. Los Transformers pueden ver todos los pedazos de un texto al mismo tiempo y entender cómo se conectan. Son muy buenos para tareas complejas de lenguaje, como traducir, resumir o responder preguntas sobre un texto. Estas redes están en el corazón de muchos modelos de lenguaje avanzado, como los que crean resúmenes o realizan traducciones instantáneas.

**6. Redes Generativas Adversarias (GANs):**

Las GANs son como un juego entre dos amigos. Uno intenta hacer dibujos de cosas como caras o paisajes y el otro trata de adivinar si son reales o falsos. Al principio, el dibujo se ve muy raro, pero cada vez que el amigo se equivoca, el que dibuja mejora su técnica. Finalmente, el dibujo se vuelve tan bueno que parece real. Las GANs son geniales para crear imágenes de cosas que no existen realmente, como paisajes imaginarios o personajes inventados.

**7. Redes Generativas:**

Estas redes crean cosas nuevas desde cero, como un pintor que inventa una obra de arte. Estas redes, muchas veces en combinación con GANs, pueden generar cualquier cosa, desde música hasta texto y obras de arte. Son perfectas para la creatividad y se usan para hacer videojuegos, canciones y hasta cuadros artísticos.

### Arquitectura más comunes para Transfer Learning

**¿Qué es el Transfer Learning:**

- Tu tienes una habilidad que ya conoces, manejas y entiendes y con esa hablidad la aplicas a una nueva tarea. La red neuronal apreande a ver ciertos tipos de imágenes (por ejemplo, gatos, perros, autos) y luego enseñarle algo nuevo (como reconocer otros animales) sin tener que empezar desde cero.

Algunas arquitecturas son:

**1. VGG:**

Esta arquitectura tiene una estructura muy simple, como una torre de piezas apiladas una sobre otra. Es útil porque cada capa ayuda a entender detalles específicos de una imagen, como los colores o los bordes.

**2. ResNet**:

En lugar de solo apilar piezas, ResNet tiene “puentes” o caminos que permiten regresar a una capa anterior para no perder ninguna información. Esto hace que sea muy buena para redes muy profundas, ayudando a que la red no olvide nada y no se confunda.



**3. Inception:**

Esta red tiene diferentes “ventanas” para ver una imagen de varias maneras al mismo tiempo. Imagina que tienes tres lupas de diferentes tamaños para ver una imagen desde diferentes ángulos, todo a la vez. Esto permite que la red capture tanto detalles pequeños como grandes de una imagen.

**4. BERT:**

Esta arquitectura es especial para trabajar con texto. En lugar de leer palabra por palabra en un orden, BERT es capaz de mirar todas las palabras juntas y entender su contexto, como si entendiera el tono y el significado de cada palabra en una oración. Es perfecto para responder preguntas, resumir textos y entender el sentido de frases complejas.

## Tercera Parte (Ejemplo):

Ejemplo práctico usando Red Neuronal Convolucional (CNN) para el reconocimiento de imágenes.

Para una empresa de moda que necesita automatizar la identificación y categorización de su inventario visual, en este caso, camisetas y pantalones. Actualmente, el proceso de clasificación de productos es manual, lo cual consume tiempo y puede resultar en errores.

Objetivo Empresarial
El objetivo de implementar este modelo es agilizar el proceso de clasificación de productos, permitiendo que el sistema identifique y etiquete automáticamente las imágenes de camisetas y pantalones. Con esto, se busca mejorar la eficiencia operativa, reducir el tiempo en tareas repetitivas y minimizar errores en la organización del inventario.

```python
# Importamos las librerías necesarias
import os
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from PIL import Image
from sklearn.preprocessing import LabelEncoder
```

```python
#Conectamos con google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
#1. Cargamos el archivo de las etiquetas
labels_df = pd.read_excel('/content/drive/MyDrive/Aprendizaje Supervisado/Bases de Datos/Etiquetas_Ropa.xlsx')
labels_df.head()
```

```python
#Separamos los nombres de los archivos y las etiquetas
filenames = labels_df.iloc[:,0].values #Nombres
labels = labels_df.iloc[:,1].values #Etiquetas
```

```python
#Codificamos las etiquetas como valores numéricos
#Instanciamos el codificador
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels) #0 para camiseta, 1 para pantalón
```

```python
#Aplicamps el one_hot encoding para tener las salidas requeridas como el Softmax
#Para las salidas de clasificación
labels_one_hot = tf.keras.utils.to_categorical(labels, num_classes=2)
```

```python
#Ajustamos el tamañoo de las imágenes para estandarizarlas y reducir la carga
img_size = (64, 64)
```

```python
#Definimos una función para caargar y procesar las imágenes
def load_and_process_image(filename, img_size):
    img = Image.open(filename)
    img = img.resize(img_size)                       #Redimensionamos la imagen
    img = np.array(img).astype('float32') / 255.0    #Normalizamos los valores a un rango de 0 a 1
    if img.shape[-1] == 4:                           #Comprobamos si existe canal alfa
      img = img[:,:,:3]
    return img
```

```python
#Cargamos las imágenes desde el sistema de archivos
#Agregamos la extensión a cada nombre de la imágen
filenames_with_extension = [fname if fname.lower().endswith(('.jpg', '.jpeg', '.png')) else fname + ".jpg" for fname in filenames]
#Definimos el directorio de las imagenes
image_dir = '/content/drive/MyDrive/Aprendizaje Supervisado/Bases de Datos/Ropa'
#Cargamos las imágenes
images = np.array([load_and_process_image(os.path.join(image_dir, fname), img_size) for fname in filenames_with_extension])
```

```python
images.shape
```

```python
#si queremos las imágenes del conjunto de datos (no las matrices numéricas)
import matplotlib.pyplot as plt
plt.imshow(images[30,:,:,:]);
```

```python
#Dividimos el conjunto de datos entrenamiento y prueba
train_images, test_images, train_labels, test_labels = train_test_split(images, labels_one_hot, test_size=0.2, random_state=42)
```

```python
from threading import active_count
#Empezamos con la construcción del modelo
model = models.Sequential ()

#Primera capa convolucional y de pooling
model.add(layers.Conv2D(32, (3,3), activation="relu", input_shape=(img_size[0], img_size[1], 3)))
model.add(layers.MaxPooling2D((2,2)))

#Segunda capa convolucional y pooling
model.add(layers.Conv2D(64, (3,3), activation="relu"))
model.add(layers.MaxPooling2D((2,2)))

#Tercera capa convolucional y pooling
model.add(layers.Conv2D(128, (3,3), activation="relu"))
model.add(layers.MaxPooling2D((2,2)))
```

|Nota en la construcción del modelo

- **Las Convolucionales:** detectan características visuales específicas de una imagen, como bordes, texturas, colores y formas básicas. Eso se logra aplicando un filtro que pasan por la imagen para encontrar patrones en secciones pequeñas de la imagen.
- Funciona como un lupa que analiza cada pequeña parte de la imagen a la vez creando capaz de características.

- **Capas de Pooling:** reducen el tamaño de los mpas de características generandps por las capas convolucionales. (Simplifica la información sin perder patrones claves).
- Funciona dividiendo el mapa de características en pequeñas regiones y seleciona solo el valor más alto de cada una. Se obtiene solo la información más destacada

```python
#Aplanamos las capad para clasificación final
model.add(layers.Flatten())
model.add(layers.Dense(64, activation="relu"))
#Salidas para cada categortía
model.add(layers.Dense(2, activation="softmax"))
```

```python
#Compilamos el modelo
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
#Entrenamiento del modelo
history = model.fit(train_images, train_labels, epochs=10, batch_size=32, validation_data=(test_images, test_labels))
```

```python
#Evaluamos el modelo
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f"Precisión en el conjunto de prueba: {test_acc}")
```

### Análisis del Modelo:

- El modelo tiene tres bloques de capas. Estas capas con para exter las características de las imagenes, entre formas, texturas y demás que le ayudan a diferenciar si es una camiseta o un pantalón.
- La precisión del modeo es del 68,75% lo que indica que el modelo no es el mejor pero tampoco es el mejor.
- Algunas de las razones por las que el modelo no puede ser más extacto es que cuando un conjunto de imágenes algo pequeño, lo cual hace que modelo no tenga sufiente información para aprender a generar diferencias.
- Puede que necesite más capas o necesite códigos más complejos en la detección de datos.
